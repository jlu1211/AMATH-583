```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('q3.csv')

plt.figure(figsize=(10,6))
plt.plot(df['n'], df['swapRows'], marker='o', label='swapRows')
plt.plot(df['n'], df['swapCols'], marker='x', label='swapCols')
plt.plot(df['n'], df['manualSwapRows'], marker='o', label='manualSwapRows')
plt.plot(df['n'], df['manualSwapCols'], marker='x', label='manualSwapCols')
plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel('Matrix size n')
plt.ylabel('Time per swap (s)')
plt.title('Row vs Column Swap Performance')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig('q3.png', dpi=150)
plt.show()
```

```python
df = pd.read_csv("q5.csv")
plt.figure(figsize=(12,6))
plt.plot(df['n'], df['GFLOPS_s'], label='GFLOPS_s')

max_n = int(df['n'].max())
pows = [2**k for k in range(int(np.log2(max_n))+1) if 2**k >= 2]
for p in pows:
    plt.axvline(x=p, linestyle=':', color='gray', linewidth=1)

plt.xticks(pows, [f"$2^{{{int(np.log2(p))}}}$" for p in pows])
plt.xlim(1, 520)
plt.xticks(fontsize=8)
plt.xlabel('Matrix size $n$')
plt.ylabel('Achieved GFLOP/s')
plt.title('Strassen Multiply Performance')
plt.grid(True, which='both', ls='--', alpha=0.3)
plt.tight_layout()
plt.savefig('q5.png', dpi=150)
plt.show()
```
